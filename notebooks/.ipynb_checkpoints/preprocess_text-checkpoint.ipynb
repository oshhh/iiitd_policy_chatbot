# Import and function definitions

```python
import re
import sys
import neuralcoref
import spacy 
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import inflect 
from nltk.stem import WordNetLemmatizer
from pyopenie import OpenIE5


p = inflect.engine() 
nlp = spacy.load('en')
neuralcoref.add_to_pipe(nlp)
stemmer= PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()

# convert number into words 
def convert_number(text): 
    temp_str = text.split() 
    new_string = [] 

    for word in temp_str: 
        if word.isdigit(): 
            temp = p.number_to_words(word) 
            new_string.append(temp) 
        else: 
            new_string.append(word) 
    temp_str = ' '.join(new_string) 
    return temp_str 

def read_raw_text(filename):
    raw_text = ''
    with open(filename) as file:
        for line in file:
            raw_text += line
    return raw_text

def coreference_resolution(text):
    return nlp(text)._.coref_resolved

def split_into_sentences(text):
    return text.replace('\n','$').split('$')

def preprocess(sentence, stop_words = [], lemmatize = False, stem = False):
    sentence = sentence.lower()
    preprocessed_sentence = ''
    for char in sentence:
        if char in 'abcdefghijklmnopqrstuvwxyz,!?.[]{}\'\"@#%^&*()-_=+\\;:<>/1234567890 ':
            preprocessed_sentence += char
    words = preprocessed_sentence.split() 
    words = [word for word in words if word not in stop_words]
    if stem:
        words = [stemmer.stem(word) for word in words]
    if lemmatize:
        words = [wordnet_lemmatizer.lemmatize(word) for word in words]
    preprocessed_sentence = ' '.join(word for word in words)
    return preprocessed_sentence
```

# Read and preprocess text

```python
# read raw text
raw_text = read_raw_text('text.txt')
# coreference resolution
resolved_text = coreference_resolution(raw_text)
# split into sentences
sentences = split_into_sentences(resolved_text)
# convert to lower case, remove uncommon special characters, stemming, lemmatisation and removing stopwords
stopwords = ['a', 'an', 'the', 'all']
preprocessed_sentences = []
for sentence in sentences:
	preprocessed_sentences.append(preprocess(sentence, stopwords))
sentences = preprocessed_sentences

sentences = [sentence for sentence in sentences if sentence]

for i in range(len(sentences)):
	print(i, sentences[i])
```

# Extract triples using OpenIE and store as edges

```python
# Get tripples from OpenIE5 for the preprocessed text
edges = []

extractor = OpenIE5('http://localhost:8000')
for s_no in range(len(sentences)):
	sentence = sentences[s_no]
	extractions = extractor.extract(sentence)
	for extraction in extractions:
		extraction = extraction['extraction']
		print(extraction['arg1']['text'], '|', extraction['rel']['text'] , '|', *[extraction['arg2s'][i]['text'] + ',' for i in range(len(extraction['arg2s']))], extraction['negated'])

		edges.append(['$' + str(s_no), extraction['rel']['text'], extraction['arg1']['text']])
		for i in range(len(extraction['arg2s'])):
			edges.append(['$' + str(s_no), 'object', extraction['arg2s'][i]['text']])
		if extraction['negated']:
			edges.append(['$' + str(s_no), 'type', 'Don\'t'])
```

# Perform coreference resolution

## Manually define canonicals and replace in text

```python
# Replace common canonicals
canonicals = {
	'hostel student': ['resident student', 'resident students', 'hostel student', 'hostel students', 'hostel resident', 'hostel residents', 'hosteller', 'hostellers'],
	'student': ['student', 'students'],
	'instructor': ['instructor', 'professor', 'faculty'],
	'on campus': ['in campus', 'inside campus', 'in iiitd', 'inside iiitd', 'in iiitd campus', 'inside iiitd campus', 'in college', 'inside college', 'on campus']
}

for edge in edges:
	for word in canonicals:
		if edge[0] in canonicals[word]:
			edge[0] = word
		if edge[2] in canonicals[word]:
			edge[2] = word
```

## Group canonicals and replace with one element of group

```python
# All entities
entities = set([])
for edge in edges:
	entities.add(edge[0])
	entities.add(edge[2])
entities = list(entities)

# Code for DSU
parent = [i for i in range(len(entities))]
size = [1 for i in range(len(entities))]

def find(x):
	p = x
	while p != parent[p]:
		p = parent[p]
	temp = x
	while temp != parent[temp]:
		t = parent[temp]
		parent[temp] = p
		temp = t
	return p 

def combine(x, y):
	x = find(x)
	y = find(y)
	if x == y:
		return
	if size[x] > size[y]:
		parent[y] = x
		size[x] += size[y]
	else:
		parent[x] = y
		size[y] += size[x]
        
# combine if word overlap is greater than 50% (using DSU)
for e1 in range(len(entities)):
	for e2 in range(len(entities)):
		words_1 = entities[e1].split()
		words_2 = entities[e2].split()
		common = 0
		for word in words_1:
			common += words_2.count(word)
		if 2 * common > 0.5 * (len(words_1) + len(words_2)):
			combine(e1, e2)


parent_of_word = {}
for e in range(len(entities)):
	parent_of_word[entities[e]] = entities[find(e)]

# replace entity by parent entity
for edge in edges:
    edge[0] = parent_of_word[edge[0]]
    edge[2] = parent_of_word[edge[2]]
```

print groups that are considered a single entity

```python
parents = {}
for word in parent_of_word:
	if parent_of_word[word] not in parents:
		parents[parent_of_word[word]] = []
	parents[parent_of_word[word]].append(word)

for parent in parents:
	print(parent)
	print(parents[parent])

print(len(entities), len(parents))
```

## Create relation "similar to" between canonicals

```python
# mark entity 1 as similar to the entity 2
for e1 in range(len(entities)):
    for e2 in range(len(entities)):
        if e1 == e2:
            continue
        words_1 = entities[e1].split()
        words_2 = entities[e2].split()
        common = 0
        for word in words_1:
            common += words_2.count(word)
        if 2 * common > 0.5 * (len(words_1) + len(words_2)):
            edges.append([entities[e1], 'similar to', entities[e2]])
```

# Plot Knowledge Graph

```python
from pyvis.network import Network
import networkx as nx
import matplotlib.pyplot as plt

nx_graph = nx.DiGraph()
nodes = {}
edge_labels = {}
for edge in edges:
	nodes[edge[0]] = '#00B2F7'
	nodes[edge[2]] = '#CAB2F7'
# nodes['Do\'s'] = '#F9BEB2'
nodes['Don\'t'] = '#F9BEB2'
node_color = []
for node in nodes:
	nx_graph.add_node(node)
	node_color.append(nodes[node])
for edge in edges:
	nx_graph.add_edge(edge[0], edge[2])
	edge_labels[(edge[0], edge[2])] = edge[1]


pos = nx.spring_layout(nx_graph)
nx.draw(nx_graph, pos, node_color = node_color, with_labels=True, font_size = 10)
nx.draw_networkx_edge_labels(nx_graph, pos, edge_labels = edge_labels, font_size = 10)

figure = plt.gcf()
figure.set_size_inches(50, 50)
plt.savefig("doc2.png", dpi=100)
```
